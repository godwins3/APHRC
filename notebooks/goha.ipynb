# CONSTRUCT RLTK DATASETS

## For HDSS

```python
class HDSS(rltk.Record):
    @property
    def id(self):
        return self.raw_object['patientid']

    @property
    def date(self):
        return self.raw_object['dob']

    @property
    def names(self):
        return list(map(lambda x: x.strip(), self.raw_object['fullname']))
```

```python
ds_hdss = rltk.Dataset(rltk.CSVReader('C:/Users/Matt/Desktop/hackathons/APHRC/data/synthetic_hdss_modified.csv'), record_class=HDSS)
```
